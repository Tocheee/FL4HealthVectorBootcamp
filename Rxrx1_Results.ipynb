```python
import os
import torch
from torch.utils.data import Dataset, random_split
from torchvision.transforms import ToTensor
from pathlib import Path
import pandas as pd

class Rxrx1Dataset(Dataset):
    def __init__(self, metadata, root, dataset_type, transform=None):
        """
        Args:
            metadata (DataFrame): A DataFrame containing image metadata.
            root (str): Root directory for the image data.
            dataset_type (str): 'train' or 'test' to specify dataset type.
            transform (callable, optional): Optional transformation to apply to the images.
        """
        self.metadata = metadata[metadata['dataset'] == dataset_type]
        self.root = root
        self.transform = transform if transform else ToTensor()
        
        self.label_map = {label: idx for idx, label in enumerate(sorted(self.metadata['sirna_id'].unique()))}
        print(self.label_map)
        self.metadata['mapped_label'] = self.metadata['sirna_id'].map(self.label_map)
        
    
    def __len__(self):
        return len(self.metadata)

    def __getitem__(self, idx):
        row = self.metadata.iloc[idx]
        experiment = row['experiment']
        plate = row['plate']
        well = row['well']
        site = row['site']
        label = row['mapped_label']  # Get the label index
        
        images = []
        for channel in range(1, 4):
            image_path = os.path.join(
                self.root, 
                f"images/{experiment}/Plate{plate}/{well}_s{site}_w{channel}.png"
            )
            image = self.load_image(image_path)
            images.append(image)
        
        concatenated_image = torch.cat(images, dim=0)
        print(concatenated_image.shape)
        return concatenated_image, label, row['sirna_id'] 

    def load_image(self, path):
        if not Path(path).exists():
            raise FileNotFoundError(f"Image not found at {path}")
        image = Image.open(path).convert('L') # Load as grayscale
        print(type(image))
        return self.transform(image)
    
    def label_frequency(self):
        """
        Prints the frequency of each label in the dataset.
        """
        label_counts = self.metadata['mapped_label'].value_counts()
        for label, count in sorted(label_counts.items()):
            original_label = [k for k, v in self.label_map.items() if v == label][0]
            print(f"Label {label} (original: {original_label}): {count} samples")
```

```python
def create_splits(dataset, train_fraction=0.8):
    """
    Splits the dataset into training and validation sets.
    
    Args:
        dataset (Dataset): The dataset to split.
        train_fraction (float): Fraction of data to use for training.
    
    Returns:
        Tuple: (train_dataset, val_dataset)
    """
    train_size = int(len(dataset) * train_fraction)
    val_size = len(dataset) - train_size
    return random_split(dataset, [train_size, val_size])
```
